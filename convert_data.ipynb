```python
#!/usr/bin/env python
# coding: utf-8
# In[ ]:
import numpy as np
import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.io import loadmat
import scipy.io
```

```python
# Load data
data = loadmat('virtual_data_10T.mat')
# Define path for each method
current_dir = os.getcwd()

NN_dir = current_dir+'/Data'
if not os.path.exists(NN_dir):
            os.makedirs(NN_dir)

# Unpack data
D=data['D']
r=data['r']
metast_indices=data['metast_indices']
x=np.squeeze(data['x'])
t=np.squeeze(data['t'])
U=data['U']
true_IC=data['phi0']

# Redefine data
data = {}
data['D'] = D
data['r'] = r
data['x'] = x
data['t'] = t
data['U'] = U
data['metast_indices'] = metast_indices
data['true_IC'] = true_IC
N = len(D)


'''
Save noiseless data
'''
# for i in list_to_train:
for i in range(0,N):
    NN_files_path = ""
    dram_files_path = ""
    patient_data = {}
    patient_data['D'] = D[i]
    patient_data['r'] = r[i]
    patient_data['x'] = x
    patient_data['t'] = t
    patient_data['U'] = U[:,:,i]
    patient_data['true_IC'] = true_IC
    patient_data['metast_index'] = metast_indices[i]
    if i < 9:
        NN_files_path = NN_path+'0'+str(i+1)+'_N00'
    else:
        NN_files_path = NN_path+str(i+1)+'_N00'
    np.save(NN_files_path,patient_data)

    
'''
Save noisy data for BINN
'''
gamma = 1.0 # proportional error constant
# 5% error
noise_level = 0.10
U_noise = U + noise_level * np.abs(U)**gamma * np.random.normal(size=U.shape)
U_noise = (U_noise>0)*U_noise

# for i in list_to_train:
for i in range(0,N):
    NN_files_path = ""
    dram_files_path = ""
    patient_data = {}
    patient_data['D'] = D[i]
    patient_data['r'] = r[i]
    patient_data['x'] = x
    patient_data['t'] = t
    patient_data['U'] = U_noise[:,:,i]
    patient_data['gamma'] = gamma
    patient_data['true_IC'] = true_IC
    patient_data['metast_index'] = metast_indices[i]
    if i < 9:
        NN_files_path = NN_path+'0'+str(i+1)+'_N10'
    else:
        NN_files_path = NN_path+str(i+1)+'_N10'
    np.save(NN_files_path,patient_data)
```
